```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img_original=cv2.imread("./peppers.jpeg",1)
```

```python
plt.imshow(img_original)
```

```python
img_rgb=cv2.cvtColor(img_original,cv2.COLOR_BGR2RGB)
img_gray=cv2.cvtColor(img_original,cv2.COLOR_BGR2GRAY)
fig=plt.figure(figsize=(10,7))
rows=1
columns=3
fig.add_subplot(rows ,columns ,1)
plt.imshow(img_original )
plt.axis("off")
plt.title("Original Image")
fig.add_subplot(rows ,columns ,2)
plt.imshow(img_rgb)
plt.axis("off")
plt.title("RGB Image")
fig.add_subplot(rows ,columns ,3)
plt.imshow(img_gray, cmap="gray")
plt.axis("off")
plt.title("Gray  Image")
```

```python
img_rgb_new_1 = cv2.cvtColor(img_original, cv2.COLOR_BGR2RGB)
```

```python
#görüntüdeki yeşil renkli nesnelerin maskelerini oluşturun (img_indexes gibi)#yeşil renk yerine kendi istediğini bir renge dönüştürün
img_rgb[img_rgb_indexes1]=[70,194,60]
img_hsv_green=cv2.cvtColor(img_original,cv2.COLOR_BGR2HSV)
lower_green=np.array([20,100,90])
upper_green=np.array([130,210,130])
mask_green=cv2.inRange(img_hsv_green,lower_green,upper_green)

fig=plt.figure(figsize=(10,7))
rows=1
columns=2

fig.add_subplot(rows ,columns ,1)
plt.imshow(img_rgb_new_1)
plt.axis("off")
plt.title("Indexes Image")
fig.add_subplot(rows ,columns ,2)
plt.imshow(mask_green, cmap="gray")
plt.axis("off")
plt.title("Original  Image")
```

```python
yellow=np.full_like(img_rgb_new_1,[255,255,0])
img_rgb_new_1[mask_green !=0]=yellow[mask_green !=0]
```

```python
#kodlarla renk oluşturma
height,width=100,100
img_red=np.zeros((height,width,3),dtype=np.uint8)
img_red[ :,:,0]=255

img_green=np.zeros((height,width,3),dtype=np.uint8)
img_green[ :,:,1]=255

img_blue=np.zeros((height,width,3),dtype=np.uint8)
img_blue[ :,:,2]=255


```

```python
fig=plt.figure(figsize=(10,7))
rows=1
columns=3

fig.add_subplot(rows ,columns ,1)
plt.imshow(img_red)
plt.axis("off")
plt.title("Red Image")
fig.add_subplot(rows ,columns ,2)
plt.imshow(img_green, cmap="gray")
plt.axis("off")
plt.title("Green  Image")
fig.add_subplot(rows ,columns ,3)
plt.imshow(img_blue)
plt.axis("off")
plt.title("Blue Image")
```

```python
img_black=np.zeros((height,width,3),dtype=np.uint8)
img_white=np.ones((height,width,3),dtype=np.uint8)*255

```

```python
fig=plt.figure(figsize=(10,7),facecolor='purple')
rows=1
columns=2

fig.add_subplot(rows ,columns ,1)
plt.imshow(img_black)
plt.axis("off")
plt.title("black Image")
fig.add_subplot(rows ,columns ,2)
plt.imshow(img_white, cmap="gray")
plt.axis("off")
plt.title("white  Image")
```
